8. Write a Program to demonstrate the usage of perceptron network.

```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.1, epochs=100):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.epochs):
            for i in range(n_samples):
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_predicted = self.activation_function(linear_output)

                update = self.learning_rate * (y[i] - y_predicted)
                self.weights += update * X[i]
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_function(linear_output)
        return y_predicted

    def activation_function(self, x):
        return np.where(x >= 0, 1, 0)

# Sample data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])  # AND gate logic

# Create and train the perceptron
perceptron = Perceptron()
perceptron.fit(X, y)

# Make predictions
predictions = perceptron.predict(X)
print("Predictions:", predictions)
```

9. Write a Program to demonstrate the usage of a Hebbian Network.

```python
import numpy as np

class HebbianNetwork:
    def __init__(self, learning_rate=0.1):
        self.learning_rate = learning_rate
        self.weights = None

    def fit(self, X):
        n_samples, n_features = X.shape
        self.weights = np.zeros((n_features, n_features))

        for i in range(n_samples):
            self.weights += self.learning_rate * np.outer(X[i], X[i])

    def predict(self, x):
        return np.dot(x, self.weights)

# Sample data (patterns)
X = np.array([[1, 1, 0, 0],
              [0, 0, 1, 1]])

# Create and train the Hebbian network
hebbian_network = HebbianNetwork()
hebbian_network.fit(X)

# Test with a noisy pattern
test_pattern = np.array([1, 0, 0, 0])  # Noisy version of the first pattern
predicted_pattern = hebbian_network.predict(test_pattern)

print("Predicted Pattern:", predicted_pattern)
```

10. write a program to implement feed forward neural network and feed backward neural network.

```python
import numpy as np

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

        # Initialize weights and biases
        self.weights_input_hidden = np.random.randn(self.input_size, self.hidden_size)
        self.bias_hidden = np.zeros((1, self.hidden_size))
        self.weights_hidden_output = np.random.randn(self.hidden_size, self.output_size)
        self.bias_output = np.zeros((1, self.output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, X):
        # Forward propagation
        self.hidden_layer_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
        self.hidden_layer_output = self.sigmoid(self.hidden_layer_input)
        self.output_layer_input = np.dot(self.hidden_layer_output, self.weights_hidden_output) + self.bias_output
        self.output = self.sigmoid(self.output_layer_input)
        return self.output

    def backward(self, X, y, learning_rate):
        # Backpropagation
        output_error = y - self.output
        output_delta = output_error * self.sigmoid_derivative(self.output)

        hidden_error = output_delta.dot(self.weights_hidden_output.T)
        hidden_delta = hidden_error * self.sigmoid_derivative(self.hidden_layer_output)

        # Update weights and biases
        self.weights_hidden_output += self.hidden_layer_output.T.dot(output_delta) * learning_rate
        self.bias_output += np.sum(output_delta, axis=0, keepdims=True) * learning_rate
        self.weights_input_hidden += X.T.dot(hidden_delta) * learning_rate
        self.bias_hidden += np.sum(hidden_delta, axis=0, keepdims=True) * learning_rate

    def train(self, X, y, epochs, learning_rate):
        for epoch in range(epochs):
            self.forward(X)
            self.backward(X, y, learning_rate)

# Example usage
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])  # XOR gate logic

nn = NeuralNetwork(input_size=2, hidden_size=2, output_size=1)
nn.train(X, y, epochs=10000, learning_rate=0.1)

print("Predictions:")
print(nn.forward(X))
```

11. Write a Program to demonstrate the usage of a CNN.

```python
import tensorflow as tf

# Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)

# Define the CNN model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=5, batch_size=32)

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print('Test accuracy:', accuracy)
```

12. Write a Program to demonstrate the usage of a RNN.

```python
import tensorflow as tf
import numpy as np

# 1. Prepare the text data
text = "This is a sample text to train the RNN."
vocab = sorted(set(text))
char_to_index = {char: index for index, char in enumerate(vocab)}
index_to_char = np.array(vocab)

# 2. Create input and target sequences
seq_length = 10
input_sequences = []
target_sequences = []

for i in range(0, len(text) - seq_length, 1):
    input_seq = text[i:i + seq_length]
    target_seq = text[i + seq_length]
    input_sequences.append([char_to_index[char] for char in input_seq])
    target_sequences.append(char_to_index[target_seq])

# 3. One-hot encode the sequences
input_sequences = tf.keras.utils.to_categorical(input_sequences, num_classes=len(vocab))
target_sequences = tf.keras.utils.to_categorical(target_sequences, num_classes=len(vocab))

# 4. Build the RNN model
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(128, input_shape=(input_sequences.shape[1], input_sequences.shape[2])),
    tf.keras.layers.Dense(len(vocab), activation='softmax')
])

# 5. Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam')

# 6. Train the model
model.fit(input_sequences, target_sequences, epochs=50)

# 7. Generate text
start_text = "This is"
for _ in range(100):
    input_seq = [char_to_index[char] for char in start_text[-seq_length:]]
    input_seq = tf.keras.utils.to_categorical(input_seq, num_classes=len(vocab))
    input_seq = input_seq[np.newaxis, ...]

    predicted_probs = model.predict(input_seq)[0]
    predicted_index = np.argmax(predicted_probs)
    predicted_char = index_to_char[predicted_index]

    start_text += predicted_char

print(start_text)
```

14. Find the cosine similarity using CBOW & Skipgram.

```python
!pip install gensim

import gensim
from gensim.models import Word2Vec
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Prepare the text data
text = "This is a sample text to train the RNN. This text is used to demonstrate CBOW and Skip-gram models."
sentences = [gensim.utils.simple_preprocess(text)]

# Train CBOW model
cbow_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, sg=0)  # sg=0 for CBOW

# Train Skip-gram model
skipgram_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, sg=1)  # sg=1 for Skip-gram

# Get word embeddings for 'text' and 'rnn'
cbow_text_embedding = cbow_model.wv['text']
cbow_rnn_embedding = cbow_model.wv['rnn']
skipgram_text_embedding = skipgram_model.wv['text']
skipgram_rnn_embedding = skipgram_model.wv['rnn']

# Calculate cosine similarity using CBOW embeddings
cbow_similarity = cosine_similarity(cbow_text_embedding.reshape(1, -1), cbow_rnn_embedding.reshape(1, -1))[0][0]

# Calculate cosine similarity using Skip-gram embeddings
skipgram_similarity = cosine_similarity(skipgram_text_embedding.reshape(1, -1), skipgram_rnn_embedding.reshape(1, -1))[0][0]

print(f"Cosine Similarity (CBOW): {cbow_similarity}")
print(f"Cosine Similarity (Skip-gram): {skipgram_similarity}")
```

15. WAP to demonstrate the concept of Positional Encoding.

```python
import numpy as np
import matplotlib.pyplot as plt

def positional_encoding(max_len, d_model):
    """
    Generates positional encoding matrix.

    Args:
        max_len: Maximum sequence length.
        d_model: Dimensionality of the model.

    Returns:
        Positional encoding matrix of shape (max_len, d_model).
    """
    pe = np.zeros((max_len, d_model))
    position = np.arange(0, max_len, dtype=np.float32).reshape(-1, 1)
    div_term = np.exp(np.arange(0, d_model, 2).astype(np.float32) * -(np.log(10000.0) / d_model))

    pe[:, 0::2] = np.sin(position * div_term)
    pe[:, 1::2] = np.cos(position * div_term)

    return pe

# Example usage:
max_len = 50  # Maximum sequence length
d_model = 512  # Dimensionality of the model

pe = positional_encoding(max_len, d_model)

# Visualize positional encoding
plt.figure(figsize=(10, 5))
plt.imshow(pe, aspect='auto', cmap='viridis')
plt.title('Positional Encoding')
plt.xlabel('Dimension')
plt.ylabel('Position')
plt.colorbar()
plt.show()
```

16. WAP to Compute the Attention Value in Attention Mechanism.

```python
import numpy as np

def compute_attention_value(query, key, value):
    """
    Computes the attention value.

    Args:
        query: Query vector (shape: [d_k]).
        key: Key vector (shape: [d_k]).
        value: Value vector (shape: [d_v]).

    Returns:
        Attention value (scalar).
    """

    # 1. Calculate the dot product between query and key
    dot_product = np.dot(query, key)

    # 2. Scale the dot product by the square root of the key dimension
    scaled_dot_product = dot_product / np.sqrt(len(key))

    # 3. Apply softmax to get attention weights
    attention_weights = np.exp(scaled_dot_product)
    attention_weights /= np.sum(np.exp(scaled_dot_product))

    # 4. Calculate the weighted sum of the values using the attention weights
    attention_value = np.dot(attention_weights, value)

    return attention_value

# Example usage:
query = np.array([1, 2, 3])
key = np.array([4, 5, 6])
value = np.array([7, 8, 9])

attention_value = compute_attention_value(query, key, value)
print(f"Attention Value: {attention_value}")
```

17. WAP to compute Attention weight & Attention output in a Transformer.

```python
import numpy as np

def scaled_dot_product_attention(query, key, value, mask=None):
    """
    Computes scaled dot-product attention.

    Args:
        query: Query tensor (shape: [batch_size, num_heads, seq_len, d_k]).
        key: Key tensor (shape: [batch_size, num_heads, seq_len, d_k]).
        value: Value tensor (shape: [batch_size, num_heads, seq_len, d_v]).
        mask: Optional mask to prevent attention to certain positions (shape: [batch_size, 1, 1, seq_len]).

    Returns:
        Tuple containing:
            - Attention output tensor (shape: [batch_size, num_heads, seq_len, d_v]).
            - Attention weights tensor (shape: [batch_size, num_heads, seq_len, seq_len]).
    """

    # 1. Calculate the dot product between query and key
    scores = np.matmul(query, key.transpose(0, 1, 3, 2))

    # 2. Scale the dot product by the square root of the key dimension
    d_k = query.shape[-1]
    scaled_scores = scores / np.sqrt(d_k)

    # 3. Apply mask (if provided)
    if mask is not None:
        scaled_scores = scaled_scores + (mask * -1e9)  # Apply mask to avoid attending to padded positions

    # 4. Apply softmax to get attention weights
    attention_weights = np.exp(scaled_scores) / np.sum(np.exp(scaled_scores), axis=-1, keepdims=True)

    # 5. Calculate the weighted sum of the values using the attention weights
    attention_output = np.matmul(attention_weights, value)

    return attention_output, attention_weights

# Example usage:
batch_size = 2
num_heads = 8
seq_len = 10
d_k = 64
d_v = 64

# Generate random tensors for query, key, and value
query = np.random.rand(batch_size, num_heads, seq_len, d_k)
key = np.random.rand(batch_size, num_heads, seq_len, d_k)
value = np.random.rand(batch_size, num_heads, seq_len, d_v)

# Compute attention
attention_output, attention_weights = scaled_dot_product_attention(query, key, value)

print(f"Attention Output Shape: {attention_output.shape}")
print(f"Attention Weights Shape: {attention_weights.shape}")
```
